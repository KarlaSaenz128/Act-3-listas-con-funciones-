# 48. cantante

```python
# 1. Definición de la clase Cantante para crear objetos con sus atributos
class Cantante:
    def __init__(self, id, nombre, genero):
        self.id = id          # Identificador único del cantante
        self.nombre = nombre  # Nombre artístico
        self.genero = genero  # Género musical (ej. Pop, Rock)

# 2. Lista global para almacenar los objetos de tipo Cantante
lista_cantantes = []

# 3. Función para agregar un nuevo cantante a la lista
def agregarCantante():
    id = input("Ingrese el ID del cantante: ")
    nombre = input("Ingrese el nombre del cantante: ")
    genero = input("Ingrese el género musical: ")
    nuevo = Cantante(id, nombre, genero)  # Se crea la instancia del objeto [3]
    lista_cantantes.append(nuevo)         # Se añade a la lista usando append() [4]
    print("Cantante agregado con éxito.")

# 4. Función para listar todos los cantantes registrados
def listarCantantes():
    if not lista_cantantes:
        print("La lista está vacía.")
    else:
        print("\n--- Lista de Cantantes ---")
        for c in lista_cantantes:         # Ciclo for para recorrer la lista [5]
            print(f"ID: {c.id} | Nombre: {c.nombre} | Género: {c.genero}")

# 5. Función para buscar un cantante por su ID
def buscarCantante():
    id_buscado = input("Ingrese el ID del cantante a buscar: ")
    for c in lista_cantantes:
        if c.id == id_buscado:            # Estructura de decisión para comparar [6]
            return c                      # Retorna el objeto si lo encuentra
    return None

# 6. Función para leer (mostrar) los datos de un cantante específico
def leerCantante():
    c = buscarCantante()
    if c:
        print(f"\nDatos encontrados:\nNombre: {c.nombre}\nGénero: {c.genero}")
    else:
        print("Cantante no encontrado.")

# 7. Función para modificar los datos de un cantante existente
def modificarCantante():
    c = buscarCantante()
    if c:
        c.nombre = input("Ingrese el nuevo nombre: ")
        c.genero = input("Ingrese el nuevo género: ")
        print("Datos actualizados correctamente.")
    else:
        print("No se pudo modificar: ID no encontrado.")

# 8. Función para borrar un cantante de la lista
def borrarCantante():
    c = buscarCantante()
    if c:
        lista_cantantes.remove(c)         # Elimina el objeto de la lista [7]
        print("Cantante eliminado.")
    else:
        print("No se pudo eliminar: ID no encontrado.")

# --- Menú Principal ---
while True:                               # Ciclo infinito para mantener el menú activo [8]
    print("\n--- MENÚ DE GESTIÓN DE CANTANTES ---")
    print("1. Agregar cantante")
    print("2. Listar cantantes")
    print("3. Leer (ver) cantante")
    print("4. Buscar cantante")
    print("5. Modificar cantante")
    print("6. Borrar cantante")
    print("7. Salir")

    opcion = input("Elija una opción (1-7): ")

    if opcion == '1':
        agregarCantante()
    elif opcion == '2':
        listarCantantes()
    elif opcion == '3':
        leerCantante()
    elif opcion == '4':
        c = buscarCantante()
        if c: print(f"El cantante {c.nombre} existe en la lista.")
        else: print("No se encontró.")
    elif opcion == '5':
        modificarCantante()
    elif opcion == '6':
        borrarCantante()
    elif opcion == '7':
        print("Saliendo del programa...")
        break                             # Rompe el ciclo para finalizar [9]
    else:
        print("Opción no válida.")
```
